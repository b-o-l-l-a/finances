```python
def assign_id(row, df):
    date = row["date"].strftime("%Y%m%d")
    amt = str(abs(row["amount"]))
    
    id = "{}_{}".format(date, amt)
    print(df.columns)
    df.where(df["id"])
    return id

#usaa_cc_df["id"] = np.nan
#usaa_cc_df["id"] = usaa_cc_df.apply(assign_id, args=(usaa_cc_df,), axis=1)
usaa_cc_df["id"] = np.nan
for idx, row in usaa_cc_df.iterrows():
    date = row["date"].strftime("%Y%m%d")
    amt = str(abs(row["amount"]))
    
    id = "{}_{}".format(date, amt)
    if id in usaa_cc_df["id"].unique():

        for i in range(2, 10):
            if "{}_{}".format(id, i) not in usaa_cc_df["id"].unique():
                id = "{}_{}".format(id, i)
                break
    
    usaa_cc_df.iloc[idx, usaa_cc_df.columns.get_loc("id")] = id
    
    
```
